# Regularization Nedir?
* Regularization modelin overfitting olmaması için coefficient tahminini 0 a yakınlaştırmaktır.
* Böylelikle model overfitting nedeniyle düzgün çalışmadığı zaman, modelin karşmaşıklığını kontrol altında tutabiliriz.
* Teknik olarak regularization, overfitting i modelin loss fonksiyonuna bir penaltı değeri ekleyerek engeller.

$$
\begin {array}{|l|}
\hline
Regularization = Loss Function + Penalty \\
\hline
\end{array}
$$

# Soru 1 a)
* Ridge regresyon ve Lasso regresyon yöntemlerinin optimizasyon problemini matematiksel olarak yazarak yöntemleri hem en küçük karelerle hem de birbirileriyle karşılaştırınız.

## Least Squares Linear Regression

* Karelerin farkının toplamını minimize etmeye çalışır.

$$
\sum_{i=1}^m (a_i^Tx - b_i)^2,
$$

*l2*-norm u olarak da bilinir.

$$\|Ax - b\|_2^2$$

* Tahmin edilen coefficient lerin her zaman positif olmasını sağlar, ama bu yüksek dimension lı verilerde overfitting ve kötü performansa neden olabilir.

## Ridge Regularization

Bu yöntem coefficient genliğinin karesi ile ilişkili bir penaltı değeri ekleyerek regularization işlemini gerçekleştirir.

$$
\begin {array}{|l|}
\hline
Ridge Regularization = Loss Function + \frac{1}{2}\lambda\sum_{j=1}^{m} w_j^2 \\
\hline
\end{array}
$$

Öyle ki :

- ${\lambda}$ : Regularization parametresi.
    * Regularizasyon un kuvvetini kontrol eder.
    * Büyük lamda daha fazla regularizasyon, modelde küçük coefficient anlamına gelir.
    * Fakat öğrenmenin çok azalmasına (underfitting) neden olabilir.
- ${w_j}$: coefficient veya her feature a denk gelen  ağırlık değerini ifade eder.
- ${m}$: feature veya coefficient sayısı

Aşağıdaki şekilde de ifade edildiğini görebiliriz;

Amacımız fonksiyonu minimize eden $\beta$ değerini bulmaktır.
$$f(\beta) = \|X\beta - Y\|_2^2 + \lambda \|\beta\|_2^2$$

## Lasso Regularizaton
Bu yöntem bazı coefficient değerlerini sıfıra çeker. Yani model bu feature u işlemlerinde ihmal (discard) eder.  
En önemsiz özellikleri ihmal etmek modelin ana özelliklerinin öne çıkmasını sağlar.

$$
\begin {array}{|l|}
\hline
Lasso Regularizaton = Loss Function + \lambda\sum_{j=1}^{m}|w_j| \\
\hline
\end{array}
$$


Aşağıdaki şekilde de ifade edildiğini görebiliriz;

Amacımız fonksiyonu minimize eden $\beta$ değerini bulmaktır.
$$f(\beta) = \|X\beta - Y\|_2^2 + \lambda \|\beta\|_1$$



## Least Squares vs Ridge vs Lasso Regularization

* Lasso least square yöntemine çok benzerdir. Tek farkı l1 form a penaltı değeri ekliyor olmasıdır.
* Lasso bazı özellikleri ihmal ederek, sparse bir modeli teşfik eder.
* Ridge coefficient değerlerinin daha stabil olmasını (genliğin karesi ile) teşfik eder.
* Bu ikisinden (Ridge veya Lasso) hangisinin seçilmesi gerektiği bilgisi veri setinin ve problemin karakteristiğine bağlıdır.

## Common Code Base

```python
import pandas as pd

# Getting data from csv file
housing = pd.read_csv('HousingData.csv')
print("housing.shape: ", housing.shape)
housing.head()
```

```python
# Check missing values
housing.isna().sum()
```

```python
# Fill missing values with column median value
for col in housing.columns:
    if housing[col].dtype != 'object':
        housing[col] = housing[col].fillna(housing[col].mean())

# Show missing value count
housing.isna().sum()
```

```python
# check if there is any duplicated values
housing.duplicated().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

print('housing.shape: ' + str(housing.shape))

# first let's define the features and target variables
X = housing.drop('MEDV', axis=1)
print('X.shape: ' + str(X.shape))

y = housing['MEDV']
print('y.shape: ' + str(y.shape))

# https://www.digitalocean.com/community/tutorials/standardscaler-function-in-python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print('X_scaled.shape: ' + str(X_scaled.shape))

# spliting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=25)

print('X_train.shape: ' + str(X_train.shape))
print('X_test.shape: ' + str(X_test.shape))
print('Y_train.shape: ' + str(Y_train.shape))
print('Y_test.shape: ' + str(Y_test.shape))
```

$$
Loss Function = \|X\beta - Y\|_2^2 \\
$$

```python
import cvxpy as cp


def loss_fn(X, Y, beta):
    return cp.norm2(cp.matmul(X, beta) - Y)**2

# NOTE: Same as above but using cp.pnorm instead of cp.norm2
# def loss_fn(X, Y, beta):
#     return cp.pnorm(X @ beta - Y, p=2)**2


def mean_square_error(X, Y, beta):
    return (1.0 / X.shape[0]) * loss_fn(X, Y, beta)   .value
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

def plot_train_test_errors(train_errors, test_errors, lambda_values, regression_name):
    plt.plot(lambda_values, train_errors, label="Train error")
    plt.plot(lambda_values, test_errors, label="Test error")
    plt.xscale("log")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xlabel(r"$\lambda$", fontsize=16)
    plt.title(regression_name + " - Mean Squared Error (MSE)")
    plt.grid(True)
    plt.show()
```

## Linear Regression

Aşağıdaki şekilde de ifade edilebilir.

$$\|Ax - b\|_2^2$$



```python
import numpy as np

m_rows, n_features = X_train.shape
print("Number of rows: ", m_rows)
print("Number of features: ", n_features)
```

```python
A = X_train
b = np.random.randn(m_rows)
x = cp.Variable(n_features)

cost = cp.sum_squares(A @ x - b)
prob = cp.Problem(cp.Minimize(cost))
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("The optimal x is")
print(x.value)
print("The norm of the residual is ", cp.norm(A @ x - b, p=2).value)
```

## Ridge Regression

$$
\begin {array}{|l|}
\hline
Ridge Regularizaton = Loss Function + \frac{1}{2}\lambda\sum_{j=1}^{m} w_j^2 \\
\hline
\end{array}
$$

Amacımız fonksiyonu minimize eden $\beta$ değerini bulmaktır.
$$f(\beta) = \|X\beta - Y\|_2^2 + \lambda \|\beta\|_2^2$$

```python
def ridge_regularizer(beta):
    return cp.pnorm(beta, p=2)**2


def ridge_objective_fn(X, Y, beta, lambdaa):
    return loss_fn(X, Y, beta) + lambdaa * ridge_regularizer(beta)
```

```python
beta = cp.Variable(n_features)
lambdaa = cp.Parameter(nonneg=True)

lambda_values = np.logspace(-2, 4, 50)
```

```python
problem = cp.Problem(cp.Minimize(ridge_objective_fn(X_train, Y_train, beta, lambdaa)))

train_errors = []
test_errors = []
ridge_beta_values = []
for v in lambda_values:
    lambdaa.value = v
    problem.solve(solver=cp.ECOS)
    train_errors.append(mean_square_error(X_train, Y_train, beta))
    test_errors.append(mean_square_error(X_test, Y_test, beta))
    ridge_beta_values.append(beta.value)
```

```python
regression_name = "Ridge Regression"

plot_train_test_errors(train_errors, test_errors, lambda_values, regression_name)
```

## Lasso Regression

$$
\begin {array}{|l|}
\hline
Lasso Regularizaton = Loss Function + \lambda\sum_{j=1}^{m}|w_j| \\
\hline
\end{array}
$$

Amacımız fonksiyonu minimize eden $\beta$ değerini bulmaktır.
$$f(\beta) = \|X\beta - Y\|_2^2 + \lambda \|\beta\|_1$$

```python
def lasso_regularizer(beta):
    return cp.norm1(beta)


def lasso_objective_fn(X, Y, beta, lambdaa):
    return loss_fn(X, Y, beta) + lambdaa * lasso_regularizer(beta)
```

```python
beta = cp.Variable(n_features)
lambdaa = cp.Parameter(nonneg=True)
problem = cp.Problem(cp.Minimize(
    lasso_objective_fn(X_train, Y_train, beta, lambdaa)))

train_errors = []
test_errors = []
lasso_beta_values = []
for v in lambda_values:
    lambdaa.value = v
    problem.solve(solver=cp.ECOS)
    train_errors.append(mean_square_error(X_train, Y_train, beta))
    test_errors.append(mean_square_error(X_test, Y_test, beta))
    lasso_beta_values.append(beta.value)
```

```python
regression_name = "Lasso Regression"

plot_train_test_errors(train_errors, test_errors, lambda_values, regression_name)
```

# SORU 3
Her bir yöntemde düzenlileştirme yolu (regularization path) çizdirerek bu grafiğin ne anlattığını ifade ediniz.

## Overfitting
*  Bir makine öğrenmesi yöntemi oluştururken dikkat etmemiz gereken en önemli noktalardan biri, modelin daha önce hiç karşılaşılmayan datalarda doğru sonuçlar oluşturabilmesini sağlamaktır.  
* Bunu sağlayabilmek için modelin training data yı ezberlemediğinden (overfitting) emin olmamız gerekir.  
* Regularization path (düzenlileştirme yolu) tekniği modelin performansını bilinmeyen datalarda geliştirmemizi sağlar. Bunu bir analog radyo alıcısında gelen gürültülü sinyalin daha temiz bir ses verecek şekilde düzeltilmesi (ince ayar) gibi düşünebiliriz.  
* Regularization path modelin önemli pattern lere odaklanmasını ve yeni datalar ile düzgün çalışabilmesini hedefler.


```python
def plot_regularization_path(lambda_values, beta_values, regression_name):
    num_coeffs = len(beta_values[0])
    print("Number of coefficients: ", num_coeffs)
    for i in range(num_coeffs):
        plt.plot(lambda_values, [wi[i] for wi in beta_values], label=housing.columns[i])
    plt.xlabel(r"$\lambda$", fontsize=16)
    plt.xscale("log")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title(regression_name + " - Regularization Path")
    plt.grid(True)
    plt.show()
```

## Ridge Regression Regularization Path

${\lambda}$: Regularization parametresi değerlerine bakarak;
* **Ridge** yönteminde ${\lambda}$'nın değeri büyüdükçe parametreler 0 a yakınlaşırlar. Fakat tam olarak sıfır olmazlar.
* Sıfıra yakınlaşma sırasına göre en son sıfıra yakınlaşan özellik RM olduğu için en önemli özelliktir diyebiliriz.

```python
regression_name = "Ridge Regression"

plot_regularization_path(lambda_values, ridge_beta_values, regression_name)
```

## Lasso Regression Regularization Path



${\lambda}$: Regularization parametresi değerlerine bakarak;
* **Lasso** yönteminde ${\lambda}$ değeri büyüdükçe parametreler 0 a doğru gider.
* ${\lambda}$'nın 100 den büyük değerleri için bazı feature ların tam olarak 0 değerine oturduğu gözlemlenmektedir.
* Grafikten okunduğu üzere en son sıfıra giden özellik RM olduğu için en önemli özellikler budur diyebiliriz.

```python
regression_name = "Lasso Regression"

plot_regularization_path(lambda_values, lasso_beta_values, regression_name)
```

# SORU 4
Yöntemlerin test ederek bir tabloda seçtiğiniz bir metrik üzerinden karşılaştırarak yorumlayınız. Hangi yöntemin çözümü seyreklik (sparsity) sağlar?


RM metriği üzerinden karşılaştırmak gerekirkse;
* Ridge regresyon yönteminde ${\lambda}$'nın 10 ile 100 değerleri arasında özelliğin değeri 0 a doğru ilerlemeye başlar.
* Lasso regresyon yönteminde neredeyse ${\lambda}$'nın 100 değerinden sonra özelliğin değeri 0 a doğru ilerlemeye başlar. Bu da Ridge ve Lasso arasında 10 katlık bir fark olduğunu gösterir.

Yukarıdaki bilgilere ek olarak sparsity hesaplaması aşağıdaki gibidir.

```python
from numpy import count_nonzero
import copy


def calculate_sparsity(beta_values, regression_name):
    print(regression_name + " beta values length: ", len(beta_values))
    print(regression_name + " beta values size: ", beta_values[0].size)

    matrix_element_count = len(beta_values) * beta_values[0].size
    print(regression_name + " matrix element count: ", str(matrix_element_count))
    print()

    zero_beta_values = matrix_element_count - count_nonzero(beta_values)
    print("Zero beta values before rounding: ", zero_beta_values)

    # Set values close to zero to zero
    epsilon = 1e-3
    for i in range(len(beta_values)):
        beta_values[i][(beta_values[i] < epsilon) & (beta_values[i] > -epsilon)] = 0

    zero_beta_values = matrix_element_count - count_nonzero(beta_values)
    print("Zero beta values after rounding: ", zero_beta_values)
    print()

    sparsity = 100 * (zero_beta_values / matrix_element_count)
    print(regression_name + " Sparsity: ", "{:.2f}".format(sparsity), "%")
```

${\lambda}$: Regularization parametresi
* **Lasso** yönteminde ${\lambda}$ değeri büyüdükçe parametreler 0 a doğru gider. ${\lambda}$'nın 10 a yakın değerlerinde parametrelerin %80 inin 0 değerine ulaştığını gözlemleriz.  
Bu da bize en önemli özelliklerin 0'a ulaşmayan en önemli özelliklerin listesini verir. Buna ek olarak 0 a çekilmiş parametreler çözümde **seyreklik (sparsity) sağlarlar**.

```python
calculate_sparsity(copy.deepcopy(lasso_beta_values), "Lasso")
```

${\lambda}$: Regularization parametresi
* **Ridge** yönteminde ise ${\lambda}$'nın değeri büyüdükçe parametreler 0 a yakınlaşırlar. Fakat tam olarak sıfır olmazlar. Bu nedenle bu yöntem **seyreklik (sparsity) sağlamaz**.

```python
calculate_sparsity(copy.deepcopy(ridge_beta_values), "Ridge")
```
